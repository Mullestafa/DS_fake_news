```python
import pandas as pd
# make a dict of words and their counts
vocab = {}

# laod the data
df = pd.read_csv('fake_news_cleaned_tokenized.csv', chunksize=10000, index_col=False)
```

```python
import ast
# iterate over the chunks
for i, chunk in enumerate(df):
    # parse as list with ast.literal_eval
    chunk['0'] = chunk['0'].apply(ast.literal_eval)
    # for every row in the chunk
    for row in chunk['0']:
        # add the words to the dict
        for word in row:
            if word in vocab:
                vocab[word] += 1
            else:
                vocab[word] = 1
        
    # print progress
    print('processed {} rows'.format((i+1)*10000))
```

```python
# sort the dict by value and save to file
import operator
sorted_vocab = sorted(vocab.items(), key=operator.itemgetter(1), reverse=True)
# save as pickle
import pickle
with open('vocab.pkl', 'wb') as f:
    pickle.dump(sorted_vocab, f)
```

```python
# convert vocab dict to dataframe
vocab_df = pd.DataFrame.from_dict(vocab, orient='index')
```

```python
vocab_df = vocab_df[vocab_df[0] > 2000]
```

```python
# load data
import pandas as pd
df = pd.read_csv('fake_news_cleaned_tokenized.csv', chunksize=10000, index_col=False)
vocab = vocab_df.to_dict()[0]
# create an hdf table where each word in the vocab is a column


# filter out words that appear less than 1000 times
for i, chunk in enumerate(df):
    # parse as list with ast.literal_eval
    chunk['0'] = chunk['0'].apply(ast.literal_eval)
    # for every row in the chunk check if words are in the vocab and appear more than 1000 times
    for row in chunk['0']:
        chunk['0'][i] = [word for word in row if word in vocab]
    
    # save to file. Append if file exists, otherwise create new file (picjke)
    if i == 0:
        chunk.to_csv('fake_news_cleaned_filtered.csv', mode='w', index=False)
    else:
        chunk.to_csv('fake_news_cleaned_filtered.csv', mode='a', header=False, index=False)
    # print progress
    print('processed {} rows'.format((i+1)*10000))
```
